### Standard Library Overview

Python's Standard Library is a vast collection of modules and packages that come bundle with Python, providing a wide range of functionalities out of the bos. Here's an overview of some of the most commonly used modules and packages in the Python Standard Library. 

```python
import array
arr=array.array('i',[1,2,3,4])
print(arr)
```

```python
import math
print(math.sqrt(16))
print(math.pi)
```

```python
## random

import random
print(random.randint(1,10))
print(random.choice(['apple','banana','cherry']))
```

```python
## Data Serialization

import json
data={'name':'Pashupati','age':22}

json_str=json.dumps(data)  ##Used to convert dictionary to json file
print(json_str)
print(type(json_str))


passed_data=json.loads(json_str) ##Convert json File to dictionary
print(passed_data)
print(type(passed_data))
```

```python
## csv

import csv 

with open('example.csv',mode='w',newline='')as file:
    writer=csv.writer(file)
    writer.writerow(['name','age'])
    writer.writerow(['Pashupati',22])

with open('example.csv',mode='r') as file:
    reader=csv.reader(file)
    for row in reader:
        print(row)
```

```python
## datetime
from datetime import datetime,timedelta

now=datetime.now()
print(now)

yesterday=now-timedelta(days=1)
print(yesterday)
```

```python
## time

import time
print(time.time())
time.sleep(2)
print(time.time())
```
