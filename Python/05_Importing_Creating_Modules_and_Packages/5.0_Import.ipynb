### Importing Modules in Python : Modules and Packages

In Python, modules and packages help organize and reuse code. Here's a comprehensive guide on how to import them.

```python
import math as m
m.sqrt(16)
```

```python
from math import sqrt,pi
print(sqrt(16))
print(sqrt(25))
print(pi)
```

```python
import numpy as np
np.array([1,2,3,4])
```

```python
from math import *  ##Import all functionss from math module
print(sqrt(16))
print(pi)
```

## Creating Custom Package

```python
from Packages.maths import addition
addition(2,3)
```
