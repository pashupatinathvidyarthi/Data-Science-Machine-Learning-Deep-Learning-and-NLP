# Introduction to Dictionaries

Dictionaries are unordered collections of items. They store data in key-value pairs. Keys must be unique and immutable (e.g., strings, numbers, or tuples), while values can be of any types.

```python
## Creating dictionaries
empty_dict={}
print(type(empty_dict))
```

```python
empty_dict=dict()
print(empty_dict)
```

```python
student={"name":"Pashupati","age":22,"grade":24}
print(student)
print(type(student))
```

```python
## Error

student={"name":"Pashupati","age":22,"name":24}  ##Do not hold duplication key pair value
print(student)
```

```python
## Accessing Dictionary Elements

student={"name":"Pashupati","age":22,"grade":'A'}
print(student)
```

```python
## Accessing Dictionary Elements
print(student['grade'])
print(student['age'])

## Accessing using get() method
print(student.get('grade'))
print(student.get('last_name'))
print(student.get('last_name',"Not Availabel"))
```

```python
## Modifying Dictionary Elements
## Dictionay are mutable, so you can add, update or delete elements
print(student)
```

```python
##Updating Value of a particular key
student["age"]=23
print(student)
```

```python
##Adding new Key Value Pair in dictionary
student["address"]="India"
print(student)
```

```python
## Deleting a Key Value Pair
print(student)
del student['age']
print(student)
```

```python
## Dictionary Methods

keys=student.keys() ##To get all the keys
print(keys)
value=student.values() ##To get all the values
print(value)

items=student.items() ##To get all key value pair in the form of list of tuple
print(items)
```

```python
##Shallow Copy

student_copy=student
print(student)
print(student_copy)
```

```python
student["name"]="Nath"
print(student)
print(student_copy)
```

```python
student_copy1=student.copy() ##Shalow Copy
print(student_copy1)
print(student)
```

```python
student["name"]="Vidyarthi"
print(student_copy1)
print(student)
```

```python
## Iterating Over a Dictionaries
## You can use loops to iterate over dictionaries, key, values or items

## Iterating over keys
for keys in student.keys():
    print(keys)
```

```python
## Iterate Over Values
for values in student.values():
    print(values)
```

```python
## Iterate over key value pairs
for key,value in student.items():
    print(f"{key}:{value}")
```

```python
## Nested Dictionaries

students={
    "student1":{"name":"Pashupati","age":22},
    "student2":{"name":"Peter","age":35}
}
print(students)
```

```python
## Access nested dictionaries elements
print(students["student2"]["name"])
print(students["student2"]["age"])
```

```python
## Iterating Over Nested Dictionaries
for student_id,student_info in students.items():
    print(f"{student_id}:{student_info}")
    for key,value in student_info.items():
        print(f"{key}:{value}")
```

```python
## Dictionary Comprehension
squares={x:x**2 for x in range(5)}
print(squares)
```

```python
## Conditional Dictionay Comprehension
evens={x:x**2 for x in range(10) if x%2==0}
print(evens)
```

# Practical Examples of Dictionary

```python
## use a dictionary to count frequency of elements in list

numbers=[1,2,2,3,3,3,4,4,4,4]
frequency={}
for number in numbers:
    if number in frequency:
        frequency[number]+=1
    else:
        frequency[number]=1
print(frequency)
```

```python
## Merge 2 dictionaries into one

dict1={"a":1,"b":2}
dict2={"b":3,"c":4}

merged_dict={**dict1,**dict2}
print(merged_dict)
```
