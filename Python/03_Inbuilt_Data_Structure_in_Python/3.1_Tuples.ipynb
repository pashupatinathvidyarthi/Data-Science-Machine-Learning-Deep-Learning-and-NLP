# Introduction to Tuples

##### Explanation:
Tuples are ordered collections of items that are immutable. They are similar to lists, but their immutability makes them different

```python
## Creating a tuple

empty_tuple=()
print(empty_tuple)
print(type(empty_tuple))
```

```python
lst=list()
print(type(lst))
tpl=tuple()
print(type(tpl))
```

```python
numbers=tuple([1,2,3,4,5,6])
numbers
```

```python
list((1,2,3,4,5,6))
```

```python
mixed_tuple=(1,"Hello World",3.14,True)
print(mixed_tuple)
```

```python
## Accessing Tuple Elements

numbers
```

```python
print(numbers[0])
print(numbers[2])
print(numbers[-1])
```

```python
numbers[0:4]
```

```python
numbers[::]
```

```python
numbers[::-1]
```

```python
## Tuples Operations

print(numbers)
print(mixed_tuple)

concatenation_tuple=numbers+mixed_tuple
print(concatenation_tuple)
```
