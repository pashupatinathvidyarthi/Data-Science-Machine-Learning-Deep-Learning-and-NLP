### Sets

Sets are built-in data type in python used to store collections of unique items. They are unordered, meaning that the elements do not follow a specific order, and they do not allow duplicate elements. Sets are useful for membership tests, eliminating duplicate entries, and performing mathematical set operations like union, intersection, difference and symmetric difference.

```python
##Creating a Set

my_set={1,2,3,4,5,6}
print(my_set)
print(type(my_set))
```

```python
my_empty_set=set()
print(my_empty_set)
```

```python
my_set1=set([1,2,3,4,5,6])
print(my_set1)
```

```python
my_set2=set([1,2,3,6,5,4,5,6]) ## Set does not hold duplicate value
print(my_set2)
```

```python
## Basic Set Operation

##Adding and removing Elements

my_set.add(7)
print(my_set)

my_set.add(7)
print(my_set)
```

```python
##Remove element from a set
my_set.remove(3)
print(my_set)
```

```python
my_set.remove(10)  ##Through an error as this element does not exist
```

```python
##Instead of upper code we write this code
my_set.discard(11)
print(my_set)
```

```python
## Pop Method
remove_element=my_set.pop()
print(remove_element)
print(my_set)
```

```python
##Clear all elements from a Set
my_set.clear()
print(my_set)
```
