### Sets

Sets are built-in data type in python used to store collections of unique items. They are unordered, meaning that the elements do not follow a specific order, and they do not allow duplicate elements. Sets are useful for membership tests, eliminating duplicate entries, and performing mathematical set operations like union, intersection, difference and symmetric difference.

```python
##Creating a Set

my_set={1,2,3,4,5}
print(my_set)
print(type(my_set))
```

```python
my_empty_set=set()
print(my_empty_set)
```

```python
my_set1=set([1,2,3,4,5,6])
print(my_set1)
```
