### Introduction to Functions

Definition:

A function is a block of code that perform a specific task. Functions helps in organizing code, reusing code. and improving readability.

```python
##Syntax

def function_name(parameters):
    "Docstring"
    ##Function Body
    return expression
```

```python
## Why functions ??

num=24
if num%2==0:
    print("The number is even.")
else:
    print("The number is odd.")
```

```python
def even_or_odd(num):
    """This Function finds even or odd number"""
    if num%2==0:
        print("The number is Even.")
    else:
        print("The number is odd.")
```

```python
## Call this Function

even_or_odd(25)
```

```python
even_or_odd(2466)
```

```python
## Function with Multiple Parameter

def add(a,b):
    "Adddition Function"
    c=a+b
    return c
```

```python
## default Parameters

def greet(name):
    print(f"Hello {name} Welcome to the Paradise")

greet("Pashupati")
```

```python
## Default Parameter
def greet(name="Guest"):  ## Default parameter for greet is "Guest" if no parameter is given in greet
    print(f"Hello {name} Welcome to the Paradise")

greet()
```

```python
## Variable Length Arguments
## Positional and Keywords Arguments


##Positional Arguments
def print_numbers(*args):   ##*args are positional arguments
    for numbers in args:
        print(numbers)
```

```python
print_numbers(1,2,3,4,5,6,7,8,"Pashupati")
```

```python
## Keywords Arguments

def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(name="Pashupati",age=22,country="India")
```

```python
def print_details(*args,**kwargs):
    for val in args:
        print(f"Positional Arguments : {val}")

    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(1,2,3,4,"Pashupati",name="Pashupati",age=22,country="India")
```
