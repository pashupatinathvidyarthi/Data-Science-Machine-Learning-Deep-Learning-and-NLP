```python
## Why functions ??

num=24
if num%2==0:
    print("The number is even.")
else:
    print("The number is odd.")
```

```python
def even_or_odd(num):
    """This Function finds even or odd number"""
    if num%2==0:
        print("The number is Even.")
    else:
        print("The number is odd.")
```

```python
## Call this Function

even_or_odd(25)
```

```python
even_or_odd(2466)
```
