### Lambda Function

Lambda Function are small anonymous functions defined using the <b>lambda</b> keyword. They can have any number of arguments but only one expression. They are commonly used for short operations or as arguments to higher-order functions. 

```python
## Syntax
lambda arguments:expression
```

```python
def addition(a,b):
    return a+b
```

```python
addition(2,3)
```

```python
addition=lambda a,b:a+b
type(addition)
print(addition(5,6))
```

```python
def even(num):
    if num%2==0:
        return True
    else:
        return False
```

```python
even(23)
```

```python
even1=lambda num:num%2==0
even1(12)
```
