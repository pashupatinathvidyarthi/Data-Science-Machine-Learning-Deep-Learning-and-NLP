```python
## Else >> The else statement executes a block of code if the condition in the if statement is false

age=16
if age>=18:
    print("You are eligible for voting")
else:
    print("Your are a minor")
```

```python
## elif 
## The elif statement allows you to check multiple conditions. It stands for 'else if'ArithmeticError

age=17

if age<13:
    print("You are a child.")
elif age<18:
    print("You are a teenager")
else:
    print("You are adult")
```
