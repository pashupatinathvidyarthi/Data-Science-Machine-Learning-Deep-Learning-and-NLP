```python
## String

str="Pashupati Nath"
for i in str:
    print(i)
```

```python
## While loop

## The while loop continues to execute as long as condition is true.

count=0

while count<5:
    print(count)
    count=count+1
```

```python
count=0

while count%2==0:
    print(count)
    count=count+1
```

```python
## Loop Control Statements

## break
## The break statemment exits the loop permaturely

##break statement

for i in range(10):
    if i==5:
        break
    print(i)
```

```python
##Continue

##The continue statements skip the current iteration and continue with the next.

for i in range(10):
    if i%2==0:
        continue
    print(i)
```

```python
## Pass
## The pass statement is a null opeartion; it does nothing

for i in range(5):
    if i%2==0:
        print("The number is",i)
        pass
    print(i)
```

```python
for i in range(5):
    if i==3:
        pass
    print(i)
```

```python
## Nested loop
## A loop inside a loop

for i in range(3):
    for j in range(2):
        print(f"i:{i} and j:{j}")
```
