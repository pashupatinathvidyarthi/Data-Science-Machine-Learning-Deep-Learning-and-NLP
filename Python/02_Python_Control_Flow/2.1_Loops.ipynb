```python
## String

str="Pashupati Nath"
for i in str:
    print(i)
```

```python
## While loop

## The while loop continues to execute as long as condition is true.

count=0

while count<5:
    print(count)
    count=count+1
```

```python
count=0

while count%2==0:
    print(count)
    count=count+1
```

```python
## Loop Control Statements

## break
## The break statemment exits the loop permaturely

##break statement

for i in range(10):
    if i==5:
        break
    print(i)
```
